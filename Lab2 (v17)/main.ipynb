```python
import numpy as np
import matplotlib.pyplot as plt

# Задання крайових кривих
def C0(u):
    return np.array([u, 0, u**2])

def C1(u):
    return np.array([u, 1, u**2])

def D0(v):
    return np.array([0, v, v**2])

def D1(v):
    return np.array([1, v, v**2])

# Задання точок на кутах параметричної області
P0 = np.array([0, 0, 0])
P1 = np.array([1, 0, 0])
P2 = np.array([0, 1, 0])
P3 = np.array([1, 1, 1])

# Функція для обчислення поверхні Кунса
def Q(u, v):
    return ((1 - v) * C0(u) + v * C1(u) + (1 - u) * D0(v) + u * D1(v)
            - ((1 - u) * (1 - v) * P0 + u * (1 - v) * P1 + (1 - u) * v * P2 + u * v * P3))

```

```python
num_points = int(input("Enter number of points: "))

# Створення сітки для параметрів u і v
u_vals = np.linspace(0, 1, num_points)
v_vals = np.linspace(0, 1, num_points)

# Створення проекцій на площини
X_projection = []
Y_projection = []
Z_projection = []

for u in u_vals:
    for v in v_vals:
        point = Q(u, v)
        # Проекція на площину x=0
        X_projection.append([point[1], point[2]])  # y і z
        
        # Проекція на площину y=0
        Y_projection.append([point[0], point[2]])  # x і z
        
        # Проекція на площину z=0
        Z_projection.append([point[0], point[1]])  # x і y

X_proj = np.array(X_projection)
Y_proj = np.array(Y_projection)
Z_proj = np.array(Z_projection)
```

```python
# Візуалізація проекцій
plt.figure(figsize=(10, 5))

# Проекція на площину x=0
plt.subplot(1, 3, 1)
plt.scatter(X_proj[:, 0], X_proj[:, 1], color='b')
plt.title("Projection on X=0")
plt.xlabel('Y')
plt.ylabel('Z')

# Проекція на площину y=0
plt.subplot(1, 3, 2)
plt.scatter(Y_proj[:, 0], Y_proj[:, 1], color='g')
plt.title("Projection on Y=0")
plt.xlabel('X')
plt.ylabel('Z')

# Проекція на площину z=0
plt.subplot(1, 3, 3)
plt.scatter(Z_proj[:, 0], Z_proj[:, 1], color='r')
plt.title("Projection on Z=0")
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()
```
